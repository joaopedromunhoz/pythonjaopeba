Problema:

Crie um programa que determine a quantidade de moedas que Marco recebeu, dado o n√∫mero total de moedas no pote e o n√∫mero de outros aventureiros no grupo.

Entrada:

A segunda linha cont√©m um n√∫mero inteiro numero_outros_aventureiros, que representa a quantidade de amigos de Marco (excluindo o pr√≥prio Marco).

(Observa√ß√£o: A regra foi alterada para o qu√°druplo, e o c√≥digo foi ajustado para usar o n√∫mero 4.)

```python

## Aula 3
total_moedas = int(input("Digite o total de moedas: "))
numero_outros_aventureiros = int(input("Digite o n√∫mero de outros aventureiros: "))

# A quantidade total de partes √©: (n√∫mero de amigos * 1 parte) + (Marco * 4 partes)
# total_moedas // (numero_outros_aventureiros + 4)
moedas_amigos = total_moedas // (numero_outros_aventureiros + 4)
moedas_marco = 4 * moedas_amigos

print(f"Marco recebeu {moedas_marco} moedas.")
```

Controle de Temperatura da Estufa
Contexto: O Centro de Pesquisa Agron√¥mica tem uma temperatura ideal de estufa di√°ria, que √© a soma da temperatura base e dos aumentos de temperatura aplicados por dois aquecedores (Aquecedor 1 e Aquecedor 2).

Voc√™ deve escrever um programa que, dada a temperatura ideal (meta) e a contribui√ß√£o conhecida de dois aquecedores, determine a temperatura aplicada pelo terceiro aquecedor (Aquecedor 3) e identifique qual dos tr√™s aquecedores contribuiu com a maior varia√ß√£o de temperatura.

Entrada:

Um n√∫mero inteiro meta_temperatura representando a temperatura ideal total.

Um n√∫mero inteiro contribuicao_aquecedor1 representando a contribui√ß√£o do Aquecedor 1.

Um n√∫mero inteiro contribuicao_aquecedor2 representando a contribui√ß√£o do Aquecedor 2.

```python
## Aula 6
def calcula_contribuicao_maior(meta_temperatura, contribuicao_aquecedor1, contribuicao_aquecedor2):
  # Calcula a contribui√ß√£o do Aquecedor 3
  contribuicao_aquecedor3 = meta_temperatura - contribuicao_aquecedor1 - contribuicao_aquecedor2

  # Identifica a maior contribui√ß√£o de temperatura entre os tr√™s aquecedores
  contribuicao_maxima = max(contribuicao_aquecedor1, contribuicao_aquecedor2, contribuicao_aquecedor3)

  return contribuicao_aquecedor3, contribuicao_maxima

# Exemplo de uso
contribuicao_aquecedor3, maior_contribuicao = calcula_contribuicao_maior(75, 20, 35)
print(f"Contribui√ß√£o do Aquecedor 3: {contribuicao_aquecedor3}¬∞C")
print(f"Maior contribui√ß√£o de temperatura entre os aquecedores: {maior_contribuicao}¬∞C")
```
